# TorsionBox

Drive the `TorsionBox` generator from a notebook. Edit the parameters, re-run the render cell, and the SVG is displayed inline.

```python
# Make the in-tree `boxes` package importable when running from notebooks/.
import sys
from pathlib import Path

REPO = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(REPO) not in sys.path:
    sys.path.insert(0, str(REPO))
```

```python
from IPython.display import SVG, display
from boxes.generators.torsionbox import TorsionBox
```

```python
# Helper: instantiate a generator, pass CLI-style args, render, write SVG.
def render(generator_cls, output="out.svg", **params):
    out = Path(output).resolve()
    argv = [f"--{k}={v}" for k, v in params.items()] + [f"--output={out}"]
    b = generator_cls()
    b.parseArgs(argv)
    b.open()
    b.render()
    data = b.close()
    out.write_bytes(data.getvalue())
    return out
```

```python
svg = render(
    TorsionBox,
    output="torsionbox.svg",
    x=300,
    y=200,
    h=40,
    nx=5,
    ny=3,
    thickness=3.0,
    skins=True,
)
display(SVG(filename=str(svg)))
```
